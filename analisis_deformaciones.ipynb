```python
import ipywidgets as widgets
from IPython.display import display
```

```python
class PerfilesExplorador:
    def __init__(self, data):
        self.data = data
        self.tipos = data['Tipo'].unique()

        self.tipo_dropdown = widgets.Dropdown(
            options=self.tipos,
            description='Tipo:'
        )

        self.designacion_output = widgets.Output()

        self.tipo_dropdown.observe(self.on_tipo_change, names='value')

        display(self.tipo_dropdown, self.designacion_output)

    def on_tipo_change(self, change):
        selected_tipo = change.new
        with self.designacion_output:
            self.designacion_output.clear_output()
            designaciones = self.data[self.data['Tipo'] == selected_tipo]['Designacion'].unique()
            if len(designaciones) > 0:
                print(f"Perfiles disponibles para familia: {selected_tipo}:")
                for designacion in designaciones:
                    print(designacion)
            else:
                print(f"No hay perfiles disponibles para familia: {selected_tipo}.")


# Create an instance of SteelProfileExplorer with your data
explorer = PerfilesExplorador(df)
```
